```
import folium
from IPython.display import display
from folium.plugins import MarkerCluster, BeautifyIcon

# Create a map centered around the world
m = folium.Map(location=[0, 0], zoom_start=2)

# Define shipping routes as a list of tuples (departure, destination, load)
shipping_routes = [
    (("LAX", 33.9416, -118.4085), ("HKG", 22.3080, 113.9185), 1000),
    (("SIN", 1.3644, 103.9915), ("JFK", 40.6413, -73.7781), 500),
    (("MIA", 25.7959, -80.2870), ("SYD", -33.8651, 151.2099), 800),
    (("AMS", 52.3080, 4.7639), ("PEK", 40.0806, 116.5849), 600)
    # Add more shipping routes here
]
# Add markers for 10 ports with cargo load
ports = [
    ("LAX", 33.9416, -118.4085, 10000),
    ("HKG", 22.3080, 113.9185, 500),
    ("SIN", 1.3644, 103.9915, 7500),
    ("JFK", 40.6413, -73.7781, 1200),
    # Add more ports here
]
marker_cluster = MarkerCluster().add_to(m)
for port in ports:
    name, lat, lon, load = port
    folium.Marker([lat, lon], popup=f"Port: {name}\nLoad: {load}").add_to(marker_cluster)
# Add shipping routes as arrows with thickness indicating cargo load
for route in shipping_routes:
    departure = route[0]
    destination = route[1]
    load = route[2]

    # Create a PolyLine between departure and destination with arrowhead
    folium.PolyLine(
        locations=[[departure[1], departure[2]], [destination[1], destination[2]]],
        weight=load/100 + 1,
        color='blue',
        tooltip=f"Load: {load}",
        arrowhead='arrow'
    ).add_to(m)
####
    # Add a triangle marker at the destination port
    folium.RegularPolygonMarker(
        location=[destination[1], destination[2]],
        number_of_sides=3,
        rotation=0,
        radius=12,
        color='red',
        fill_color='red'
    ).add_to(m)
# Display the map
display(m)
```

```
#0518: output visualized map of shipment with direction and quantity
import folium
from IPython.display import display
from folium.plugins import MarkerCluster, BeautifyIcon

# Create a map centered around the world
m = folium.Map(location=[0, 0], zoom_start=2)

# Define shipping routes as a list of tuples (departure, destination, load)
shipping_routes = [
    (("LAX", 33.9416, -118.4085), ("HKG", 22.3080, 113.9185), 1000),
    (("SIN", 1.3644, 103.9915), ("JFK", 40.6413, -73.7781), 5000),
    (("MIA", 25.7959, -80.2870), ("SYD", -33.8651, 151.2099), 9800),
    (("AMS", 52.3080, 4.7639), ("PEK", 40.0806, 116.5849), 600)
]

# Add markers for ports with cargo load
marker_cluster = MarkerCluster().add_to(m)
for route in shipping_routes:
    departure = route[0]
    destination = route[1]
    load = route[2]

    # Add departure port marker with port name and load in popup
    folium.Marker(
        [departure[1], departure[2]],
        popup=f"Port: {departure[0]}\nLoad: {load}",
        icon=BeautifyIcon(
            number=departure[0],
            iconShape='marker',
            borderColor='black',
            textColor='black',
            innerIconStyle='margin-top:0px;'
        )
    ).add_to(marker_cluster)

    # Add destination port marker with port name in popup
    folium.Marker(
        [destination[1], destination[2]],
        popup=f"Port: {destination[0]}",
        icon=BeautifyIcon(
            number=destination[0],
            iconShape='marker',
            borderColor='black',
            textColor='black',
            innerIconStyle='margin-top:0px;'
        )
    ).add_to(marker_cluster)

# Add shipping routes as arrows with thickness indicating cargo load
for route in shipping_routes:
    departure = route[0]
    destination = route[1]
    load = route[2]

    # Create a PolyLine between departure and destination with arrowhead
    folium.PolyLine(
        locations=[[departure[1], departure[2]], [destination[1], destination[2]]],
        weight=load / 500 + 1,
        color='blue',
        tooltip=f"Load: {load}",
        arrowhead='arrow'
    ).add_to(m)
    # Add a triangle marker at the destination port
    folium.RegularPolygonMarker(
        location=[destination[1], destination[2]],
        number_of_sides=3,
        rotation=1,
        radius=16,
        color='red',
        fill_color='red', fill_opacity=7
    ).add_to(m)
# Display the map
display(m)
```
